Plotting different fule usages against GHGEmissionsIntensity(kgCO2e/ft2)

```python
fig, (ax1, ax2, ax3 , ax4 , ax5, ax6 , ax7) = plt.subplots(1, 7,figsize=(30,5))
ax1.scatter(df_2015['SiteEUI(kBtu/sf)'],df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='red',alpha=0.25)
ax1.set_xlabel('SiteEUI(kBtu/sf)')
ax1.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax2.scatter(df_2015['SourceEUI(kBtu/sf)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='blue',alpha=0.25)
ax2.set_xlabel('SourceEUI(kBtu/sf)')
ax2.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax3.scatter(df_2015['SiteEnergyUse(kBtu)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='green',alpha=0.25)
ax3.set_xlabel('SiteEnergyUse(kBtu)')
ax3.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax4.scatter(df_2015['SteamUse(kBtu)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='pink',alpha=0.25)
ax4.set_xlabel('SteamUse(kBtu)')
ax4.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax5.scatter(df_2015['Electricity(kBtu)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='grey',alpha=0.25)
ax5.set_xlabel('Electricity(kBtu)')
ax5.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax6.scatter(df_2015['NaturalGas(kBtu)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='purple',alpha=0.25)
ax6.set_xlabel('NaturalGas(kBtu)')
ax6.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')

ax7.scatter(df_2015['OtherFuelUse(kBtu)'], df_2015['GHGEmissionsIntensity(kgCO2e/ft2)'],color='orange',alpha=0.25)
ax7.set_xlabel('OtherFuelUse(kBtu)')
ax7.set_ylabel('GHGEmissionsIntensity(kgCO2e/ft2)')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,8))
plt.title('GHG Emissions Intensity(kgCO2e/ft2) for all Building Types')
sns.barplot(x='BuildingType',y='GHGEmissionsIntensity(kgCO2e/ft2)',data=df_2015,palette='rainbow')
```

From the above plot it is evident that maximum Emission Intensity is from Campus Building Type

```python
plt.figure(figsize=(20,10))
plt.title('GHG Emissions Intensity(kgCO2e/ft2) for all Neighbourhoods for the year 2015')
sns.barplot(x='Neighborhood',y='GHGEmissionsIntensity(kgCO2e/ft2)',data=df_2015,palette='pastel')
```

```python
plt.figure(figsize=(20,10))
plt.title('Energy Star Score for Every neighbourhood for the year 2015')
sns.lineplot(x='Neighborhood',y='ENERGYSTARScore',data=df_2015,color='red')
```

```python
plt.figure(figsize=(20,10))
plt.title('Energy Star Score for Every neighbourhood for the year 2016')
sns.lineplot(x='Neighborhood',y='ENERGYSTARScore',data=df_2016)
```

```python
plt.figure(figsize=(20,15))
plt.title('Heatmap for the year 2015')
sns.heatmap(data=df_2015.corr(),annot=True, cmap='YlGnBu')
```
